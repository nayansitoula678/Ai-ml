```python
'''
open() # to open file
write() to create a file and write data into it
read() to read data from a file
close() to close the file after performing operations       

'''
#text file operations
#csv,excel-> pandas
'''
r->used for reading the file
w -> used for writing into the file 
'''
```

```python
#creating a text file
#open(file_name,operation)
file=open("abc.txt","w") # create a file and open it in write mode
#adding content to the file
content="Hello, World!"
file.write(content)

file.close()
```

```python
file=open("abc.txt","r") # open the file in read mode
content=file.read() # read the content of the file
print(content)
file.close()
```
