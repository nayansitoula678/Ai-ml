```python
#greet function
def greet():
    print("Hello, welcome to the world of Python programming!")
greet() # call back function
```

```python
def greet_name(name):
    print(f'Hello, {name}')
     

greet_name('Alice') # call back function with argument    
   
```

```python
#create a function to print personal details of a person
def personal_details(name="unknown Name", age=0, city="unknown", qualification="NA"):
    print(f'Name: {name}')
    print(f'Age: {age}')
    print(f'City: {city}')
    print(f'qualification: {qualification}')
    
```

```python
personal_details('Alice', 30, 'New York')
personal_details('Bob', 25, 'Los Angeles', 'Bachelor\'s Degree')
```

```python
personal_details()
```

```python
#error handling
#try,except(catch),else,finally
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    result = num1 / num2
except ValueError:
    print("Please enter valid integers.")
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
else:
    print(f"Result: {result}")
finally:
    print("Execution completed.")
```
