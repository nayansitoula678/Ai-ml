```python
import pandas as pd

path = 'Data/Data/HR_Dataset Refresh.csv'
hr = pd.read_csv(path)
hr.head(2)
```

```python
#feature engineering - creating new features from existing features
'''
feature engineering is the process of using domain knowledge to create new features from existing data. It can help improve the performance of machine learning models by providing them with more relevant information.
for example, if we have a dataset of employee information, we can create a new feature called "tenure" by calculating the difference between the current date and the employee's hire date. This new feature can provide valuable insights into employee retention and turnover.
extracting months year and a dayfrom column
'''
```

```python
#.dtype is used to check data type of a column
hr['DateofHire'].dtype
```

```python
#convert dateofhire to datetime datatype
hr['DateofHire'] = pd.to_datetime(hr['DateofHire'])
```

```python
#extract year from date
hr['YearofHire'] = hr['DateofHire'].dt.year
```

```python
hr
```

```python
hr['Hire_Month']=hr['DateofHire'].dt.month
hr['Hire_Day']=hr['DateofHire'].dt.day  
```

```python
hr
```

```python
hr['Hire_Month_Name']=hr['DateofHire'].dt.month_name()
hr['Hire_Day_Name']=hr['DateofHire'].dt.day_name()
```

```python
hr['Department']=hr['Department'].str.strip()  #remove leading and trailing spaces
```

```python
hr['Gender']=hr['Gender'].str.strip()  #remove leading and trailing spaces  
```

```python
#data analysis
#Groupby and aggregate functions are used while data analysis using python pandas
#GroupBy-> aggregate funtions compulsary
#aggregate functions-> Group by not necessary

```

```python
#find total salary of employee associate with each department
'''
syntax:
emp_sal=dataframe.groupby(['column_name1','column_name2'])['column_name3'].agg('aggregate_function')
single grouby
emp_sal=dataframe.groupby('column_name2')['aggcolumn'].sum()

'''
```

```python
#find total salary of employee associate with each deparment
emp_sal=hr.groupby('Department')['Salary'].sum().reset_index()
emp_sal
```

```python
#find total number of employeees who works in each department
emp_dep= hr.groupby('Department')['EmpID'].count().reset_index()
emp_dep

```

```python
emp_dep= hr.groupby('Department')['EmpID'].count().reset_index().rename(
columns={'EmpID':'Total_Employees'}
)
emp_dep
```

```python
emp_dep= hr.groupby('Department').agg(
Total_Employees=(    'EmpID','count'),Total_Salary=('Salary','sum'),
Avg_salary=('Salary','mean')

).reset_index(

)
emp_dep
```

```python
#find details of deparment which has less than 40 employees
emp_dep[emp_dep['Total_Employees']<40]  


```

```python
emp_dep['Avg_salary']=emp_dep['Avg_salary'].map("Rs.{:.2f}".format)
emp_dep
```

```python
emp_dep['Total_Salary']=emp_dep['Total_Salary'].map("Rs.{:.2f}".format)
emp_dep
```
